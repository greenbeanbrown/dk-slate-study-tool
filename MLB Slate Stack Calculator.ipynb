```python
# Need a function that transforms the Count into a string
```

```python
def convert_stack_to_string(mlb_stacks_df):
    
    # Convert the stack count to a single string, delimited by dashes
    stack_string = str('-'.join(str(int(x)) for x in mlb_stacks_df['Count']))
    
    return(stack_string)
```

```python
all_stacks = [convert_stack_to_string(lineup_stack) for lineup_stack in list_of_all_lineup_stacks]

```
